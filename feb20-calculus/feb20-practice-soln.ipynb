## Solution 1

The only real trick here is remembering to take derivatives and watching out for the $p = 0$ case. We're pretty cavalier about how we handle zeros here, because in this case it really doesn't matter.

The other thing to note is that we're doing gradient _ascent_ (we're maximizing something), so watch your signs. You should confirm on your own that if you negate the objective function and run gradient descent, you end up getting the right thing.

```python
p = np.linspace(0, 1, 100)
```

```python
entropy = -p * np.log(p) - (1-p)*np.log(1-p)
```

```python
plot(p, entropy)
```

```python
d_entropy = -(np.log(p) + 1 - 1 /(1 - p) - np.log(1-p) + p / (1-p))

def gr_ent(p):
    if p == 0: 
        return 0.0 
    else: 
        return -(np.log(p) + 1 - 1 /(1 - p) - np.log(1-p) + p / (1-p))
```

```python
plot(p, d_entropy)
```

```python
w = 0.1
gr = gr_ent(w)

alpha = 0.01

while np.abs(gr) > 0.0001:
    w = w + alpha * gr
    gr = gr_ent(w)
    
print(w)
```

## Problem 2

Once again, there isn't much actually going on here. You're given a set $\{(x_i, y_i)\}_{i=1}^n$ of points and you need to find the $a$ and $b$ that minimize a function. In this case it just so happens that the function has two special properties: 

1. it's quadratic, so it has a single global minimum.
2. It just so happens that the function computes the error from approximating your data via the function $f(x) = ax + b$. 

We care much more about the first point than the second, though if you plot the points in your set you should see that they form (for the examples above) a line.

In addition to computing the optimal parameters for a line, I also plot some contours of the error function and show the sequence of points visited by gradient descent in the process of minimizing the function. You can clearly see the quadratic nature of the error function via the fact that the isocontours are ellipses. You can also see how gradient descent behaves as you move from a random starting point to the function's global optimum.

```python
scatter(x1,y1)
```

```python
def lr_objective(x,y,a,b):
    return (1.0 / (2 * x.shape[0])) * np.sum( ((a * x + b) - y)**2 )
```

```python
def grad_lr(x, y, a, b):
    df_x = (1.0 / x.shape[0]) * np.sum(((a * x + b) - y)*x)
    df_y = (1.0 / y.shape[0]) *  np.sum((a * x + b) - y)
    return np.asarray([df_x, df_y])
```

```python
start = 25 * np.random.randn(2)
alpha = 0.1

x = x1
y = y1

w = start
gr = grad_lr(x, y, w[0], w[1])
history = []
while norm(gr) > 0.001:
    history.append(w)
    w = w - alpha * grad_lr(x,y,w[0], w[1])
    gr = grad_lr(x,y, w[0], w[1])
    
a1 = np.linspace(-50, 50, 100)
b1 = np.linspace(-50, 50, 100)

A,B = np.meshgrid(a1, b1)

z = np.zeros((A.shape[0], B.shape[0]))

for a in range(A.shape[0]):
    for b in range(B.shape[0]):
        z[a,b] = lr_objective(x,y, A[a,b], B[a,b])

contour(A,B, z, 32)
    
for pt in history:
    scatter(pt[0], pt[1])
    
print w
```

In the case where we optimize `x1` and `y1`, notice that gradient descent very quickly moves to the "northeast" until it gets to the major axis of the innermost ellipse. It then takes many steps (many more than it should) to get to the center of that ellipse. This should tell you something about how gradient descent behaves: if the error surface has these long "valleys," then gradient descent will take a long time to meander through them... Since the matrix of second partial derivatives (the Hessian) tells us about the shape of the error surface, you should think about how the eigenvalues of the Hessian tell you about the behavior of the algorithm.
