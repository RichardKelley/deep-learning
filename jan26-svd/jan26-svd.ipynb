```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt 
```

```python
# set the seed for repeatability
numpy.random.seed(12345)
size = 25
data = np.asarray([[x, np.random.normal(x, 0.1)] for x in np.linspace(-2, 2, size)])
```

```python
# normalize the data
#data = data - np.mean(data, axis=0)
data = data / np.std(data, axis=0)
```

```python
# visualize our data
scatter(data[:, 0], data[:, 1],)
```

```python
np.dot(data.T, data)/ size
```

```python
# Compute sigma - notice that this is the formula for covariance...
sigma = (1.0 / size) * np.sum([np.outer(data[i], data[i]) for i in range(size)], axis=0)
```

```python
# What does sigma look like?
sigma
```

### New Stuff: Singular Value Decomposition (SVD)

Given a matrix $A \in \mathbb{R}^{n \times n}$, we say that $A$ is _symmetric_ if $A^T = A$. We say that a matrix $A \in \mathbb{R}^{n \times n}$ is _positive semidefinite_ if $\forall x \in \mathbb{R}^n, x^T A x \ge 0$.

The _singular value decomposition_ of an $m \times n$ real matrix $A$ consists of three matrices $U$, $S$, and $V^T$, satisfying the relationship $A = U S V^T$. In this case, the matrix $U$ is orthogonal, $S$ is a diagonal matrix whose nonzero elements are called the _singular values_ of $A$, and $V^T$ is an orthogonal matrix. Make sure you can figure out the dimensions of these matrices - this should not be hard.

When $A$ is symmetric positive semidefinite, the SVD of $A$ has a neat property: The entries of $S$ are the eigenvalues of $A$, and the columns of $U$ are the corresponding eigenvectors. It is almost always the case that the entries of $S$ are listed in decreasing order. 

When $A$ is a covariance matrix, it will always be symmetric positive semidefinite, so in our present example we can use `np.linalg.svd` instead of `np.linalg.eig` to get the eigenvalues and eigenvectors of `sigma`. This is primarily a convenience, but we'll see examples where SVD is used even when eigendecomposition wouldn't make sense.

```python
u,s,v = np.linalg.svd(sigma)
```

### End New Stuff

```python
# Apply the inverse of U to the data
data_rot = np.asarray([np.dot(u, data[i]) for i in range(size)])
```

```python
# Plot the rotated data - NOTICE THE SCALING AND ASPECT RATIO ISSUE
scatter(data_rot[:,0], data_rot[:, 1])
```

```python
# Plot everything up to the rotation
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,2,1)
ax.autoscale(enable=False)
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
ax.scatter(data[:,0], data[:,1])
ax = fig.add_subplot(1,2,2)
ax.scatter(data_rot[:,0], data_rot[:,1])
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
```

```python
# Deep copy of the rotated data
data_reduced = data_rot.copy()
```

```python
# We want a line, but we want it in 2-space
data_reduced[:,1] = 0
```

```python
# Plot everything up to the 1D projection
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,3,1)
ax.autoscale(enable=False)
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.scatter(data[:,0], data[:,1])
ax = fig.add_subplot(1,3,2)
ax.scatter(data_rot[:,0], data_rot[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax = fig.add_subplot(1,3,3)
ax.scatter(data_reduced[:,0], data_reduced[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
```

```python
# Deep copy of the reduced data
data_approx = data_reduced.copy()
```

```python
# Invert the rotation
data_approx = np.asarray([np.dot(v, data_approx[i]) for i in range(size)])
```

```python
# Plot everything
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(2,2,1)
ax.autoscale(enable=False)
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.scatter(data[:,0], data[:,1])
ax = fig.add_subplot(2,2,2)
ax.scatter(data_rot[:,0], data_rot[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax = fig.add_subplot(2,2,3)
ax.scatter(data_reduced[:,0], data_reduced[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax = fig.add_subplot(2,2,4)
ax.scatter(data_approx[:,0], data_approx[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
```
