## A Tougher Example

Let's look at an example with a single global optimum that happens to be very hard to find.

```python
x_start, x_end = -4, 4
y_start, y_end = -4, 4

X1 = np.linspace(x_start, x_end, 100)
Y1 = np.linspace(y_start, y_end, 100)

x,y = np.meshgrid(X1, Y1)

z = np.where((x-1.5)**2 + (y+1.5)**2 > 1, np.sin(x) * np.sin(y),  np.sin(x) * np.sin(y) - 0.5)
```

```python
xlim(-4, 4)
ylim(-4, 4)
contour(x,y, z, 64)
```

```python
# compute the value of the gradient vector at a point:
grad_z = lambda x,y : np.asarray([np.cos(x) * np.sin(y), np.sin(x) * np.cos(y)])
```

```python
def gradient_descent(w_start, alpha, gradient, steps = 100):
    w = w_start
    gr = gradient(w[0], w[1])
    history = []
    while norm(gr) > 0.1 and steps > 0:
        history.append(w)
        w = w - alpha * gradient(w[0], w[1])
        gr = gradient(w[0], w[1])
        steps -= 1
        
    xlim(-4, 4)
    ylim(-4, 4)
    contour(x,y, z, 64)
    
    for pt in history:
        scatter(pt[0], pt[1])
        
    return w
```

```python
gradient_descent([0.5,-1], 0.5, grad_z)
```

```python
gradient_descent([-1,2], 0.5, grad_z)
```

```python
gradient_descent([1.5,0.5], 0.5, grad_z)
```

```python
gradient_descent([1,2], 0.5, grad_z)
```

```python
gradient_descent([-1,-1], 0.5, grad_z)
```
