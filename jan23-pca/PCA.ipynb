```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt 
```

We're going to start by generating `size` data points in the plane. To show a case where PCA really shines, we're going to generate data that more or less sits on a line, so that projecting to 1D really makes sense. We'll do this by starting with the line $y = 5x$ and adding some normally distributed noise.

```python
# set the seed for repeatability
numpy.random.seed(12345)
size = 25
data = np.asarray([[x, np.random.normal(x, 0.1)] for x in np.linspace(-2, 2, size)])
```

It's often the case in the real world that different features of the data (components of the vector) have different scales. We want our data to have the same scale, so we'll normalize it to have zero mean and unit variance. This is not necessary for some forms of data (natural images and audio are two examples).

```python
# normalize the data
data = data - np.mean(data, axis=0)
data = data / np.std(data, axis=0)
```

```python
# visualize our data
scatter(data[:, 0], data[:, 1],)
```

Once our data is normalized, we need to compute the following matrix: $$\Sigma = \frac{1}{m} \sum_{i=1}^m x_i x_i^{T}.$$ If the data has zero mean then this is just the covariance of the data.

If the `axis = 0` part is confusing, check out the numpy documentation.

```python
# Compute sigma - notice that this is the formula for covariance...
sigma = (1.0 / size) * np.sum([np.outer(data[i], data[i]) for i in range(size)], axis=0)
```

```python
# What does sigma look like?
sigma
```

The `np.linalg.eig` function returns the eigenvalues `v` and eigenvectors `U` of `sigma`. The eigenvalues are not necessarily ordered, and the eigenvectors are returned in a matrix such that the $i$th column of `U` corresponds to `v[i]`.

```python
# Compute the eigenvectors and eigenvalues
v,U = np.linalg.eig(sigma)
```

```python
# What do the eigenvectors and eigenvalues look like?
print(v)
print(U)
```

The `eig` function will return the eigenvalues and eigenvectors of $\Sigma$, but it doesn't necessarily order things the way we want - we want the first column of `U` to contain the "principal eigenvector," which is just the eigenvector corresponding to the largest eigenvalue. So we use some numpy magic to rearrange the columns...

```python
# Make sure the columns are sorted correctly
U[:,np.argsort(v)[::-1]]
```

A matrix having the property below (that $U * U^T = I$) is called _orthogonal_. A neat property of orthogonal matrices is that multiplying an orthogonal matrix with a vector corresponds to a rotation of that vector about some point. We're taking advantage of that here.

```python
# Confirm that inv(U) == U.T
np.dot(U, U.T)
```

Some numpy trickery: matrix-vector multiply is implemented via `np.dot`.

```python
# Apply the inverse of U to the data
data_rot = np.asarray([np.dot(U.T, data[i]) for i in range(size)])
```

```python
# What does the data look like?
data_rot[0]
```

```python
# Plot the rotated data - NOTICE THE SCALING AND ASPECT RATIO ISSUE
scatter(data_rot[:,0], data_rot[:, 1])
```

```python
# What does a data point look like?
data[1]
```

```python
# Plot everything up to the rotation
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,2,1)
ax.autoscale(enable=False)
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
ax.scatter(data[:,0], data[:,1])
ax = fig.add_subplot(1,2,2)
ax.scatter(data_rot[:,0], data_rot[:,1])
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
```

```python
# Deep copy of the rotated data
data_reduced = data_rot.copy()
```

```python
# We want a line, but we want it in 2-space
data_reduced[:,1] = 0
```

```python
# Plot everything up to the 1D projection
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,3,1)
ax.autoscale(enable=False)
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.scatter(data[:,0], data[:,1])
ax = fig.add_subplot(1,3,2)
ax.scatter(data_rot[:,0], data_rot[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax = fig.add_subplot(1,3,3)
ax.scatter(data_reduced[:,0], data_reduced[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
```

```python
# Deep copy of the reduced data
data_approx = data_reduced.copy()
```

```python
# Invert the rotation
data_approx = np.asarray([np.dot(U, data_approx[i]) for i in range(size)])
```

```python
# Plot everything
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(2,2,1)
ax.autoscale(enable=False)
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.scatter(data[:,0], data[:,1])
ax = fig.add_subplot(2,2,2)
ax.scatter(data_rot[:,0], data_rot[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax = fig.add_subplot(2,2,3)
ax.scatter(data_reduced[:,0], data_reduced[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax = fig.add_subplot(2,2,4)
ax.scatter(data_approx[:,0], data_approx[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
```

Suppose we have eigenvalues $\lambda_1, \ldots, \lambda_n$. Then a good rule of thumb for estimating the "amount of variance" that we retain by using $k$ dimensions instead of $n$ is $$\frac{\sum_{i=1}^k \lambda_i}{\sum_{i=1}^n \lambda_i}.$$

In general, the more variance you can retain, the better. This gets harder in spaces of much higher dimensionality, but a good rule of thumb is to choose $k$ such that 99% of the variance can be retained.

```python
# How much of the variance are we retaining?
v[0] / (v[0] + v[1])
```
