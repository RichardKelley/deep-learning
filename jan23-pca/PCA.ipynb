```python
import numpy as np
import sklearn
import matplotlib.pyplot as plt 
```

```python
# set the seed for repeatability
numpy.random.seed(12345)
size = 20
data = np.asarray([[x, np.random.normal(5 * x, 1)] for x in np.linspace(-2, 2, size)])
```

```python
# normalize the data
data = data - np.mean(data, axis=0)
data = data / np.std(data, axis=0)
```

```python
# visualize
scatter(data[:, 0], data[:, 1],)
```

```python
sigma = (1.0 / size) * np.sum([np.outer(data[i], data[i]) for i in range(size)], axis=0)
```

```python
sigma
```

```python
v,U = np.linalg.eig(sigma)
```

```python
v,U
```

```python
U[:,np.argsort(v)[::-1]]
```

```python
np.dot(U, U.T)
```

```python
data_rot = np.asarray([np.dot(U.T, data[i]) for i in range(size)])
```

```python
data_rot[0]
```

```python
scatter(data_rot[:,0], data_rot[:, 1])
```

```python
data[1]
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,2,1)
ax.autoscale(enable=False)
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
ax.scatter(data[:,0], data[:,1])
ax = fig.add_subplot(1,2,2)
ax.scatter(data_rot[:,0], data_rot[:,1])
ax.set_xlim([-2,2])
ax.set_ylim([-2,2])
```

```python
data_reduced = data_rot.copy()
```

```python
data_reduced[:,1] = 0
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,3,1)
ax.autoscale(enable=False)
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.scatter(data[:,0], data[:,1])
ax = fig.add_subplot(1,3,2)
ax.scatter(data_rot[:,0], data_rot[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax = fig.add_subplot(1,3,3)
ax.scatter(data_reduced[:,0], data_reduced[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
```

```python
data_approx = data_reduced.copy()
```

```python
data_approx = np.asarray([np.dot(U, data_approx[i]) for i in range(size)])
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(2,2,1)
ax.autoscale(enable=False)
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax.scatter(data[:,0], data[:,1])
ax = fig.add_subplot(2,2,2)
ax.scatter(data_rot[:,0], data_rot[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax = fig.add_subplot(2,2,3)
ax.scatter(data_reduced[:,0], data_reduced[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
ax = fig.add_subplot(2,2,4)
ax.scatter(data_approx[:,0], data_approx[:,1])
ax.set_xlim([-4,4])
ax.set_ylim([-4,4])
```
